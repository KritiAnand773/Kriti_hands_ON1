```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url='/Shopping_data.csv'
df=pd.read_csv(url)
print(df)
```

```python
mean_value=df['Age'].mean()
print(f"mean_value:-{mean_value}")
```

```python
median_value=df['Age'].median()
print(f"median_value:-{median_value}")
```

```python
skew_value=df['Age'].skew()
print(f"Standerd deviation value:-{skew_value}")
```

```python
kurt_value=df['Age'].kurt()
print(f"Kurotosis value:-{kurt_value}")
```

```python
import matplotlib.pyplot as plt
df['Age'].plot(kind='hist', title='Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
#bar plot for the number of Annual income
df['Annual Income (k$)'].value_counts().plot(kind='bar',title='Annual Income (k$)')
plt.xlabel('Annual Income (k$)')
plt.show()
```

```python
#history of Age
plt.hist(df['Age'], bins=30,color='blue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box plot for item Age
plt.boxplot(df['Age'], vert=False)
plt.title('Box plot of Age')
plt.show()
```
